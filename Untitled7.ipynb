```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('vacancies.xlsx', header=0, index_col='#') 
# Maosh qismidagi sonlar string va ularni floatga o'tkazamiz
def change_value(numb):
    numb = str(numb)
    numb = numb.replace('.', '')
    numb = numb.replace(',','.')
    return int(numb)/100
df['Маош'] = df['Маош'].apply(change_value)
```

Berilganlar orasidan eng yuqori maosh taklif etuvchini topamiz

```python
maximum_maosh = df[df['Маош']==df['Маош'].max()]
maximum_maosh 
```

```python
# O'rtacha maosh == Average
average_salary = df['Маош'].mean()
average_salary
```

```python
regi = df.groupby('Вилоят')
regi
```

```python
regi['Маош'].describe()['mean']
```

```python
def remove(x):
    return x.split()[0]
```

```python
# Hududlar bo'yicha o'rtacha ish haqi
plt.figure(figsize=(15,8))
plt.bar(list(map(remove, regi['Вилоят'].describe()['top'])), regi['Маош'].describe()['mean'])
plt.show()
```

```python
# Hududlar bo'yicha maksimal ish haqi
plt.figure(figsize=(15,8))
plt.bar(list(map(remove, regi['Вилоят'].describe()['top'])), regi['Маош'].describe()['max'])
plt.show()
```

```python
# Hududlar bo'yicha eng past ish haqi
plt.figure(figsize=(15,8))
plt.bar(list(map(remove, regi['Вилоят'].describe()['top'])), regi['Маош'].describe()['min'])
plt.show()
```

```python
# Indexlashda xato kelib chiqayotgan ekan, shu sabab reset_index() qilindi
dff = df.reset_index()
```

```python
# Mutaxassisliklarni ajratib olamiz
specialities = df['Мутахассислик']
specialities.drop_duplicates(inplace=True)
```

```python
# Mutaxassisliklar va ulargaa ajratilgan o'rtacha maoshlar dict qilib olamiz
salary_with_specs = {}
for i in specialities:
    salary_with_specs[i] = dff[dff['Мутахассислик'] == i]['Маош'].mean()
    
salary_with_specs
```

```python
len(salary_with_specs)
```

```python
# Afsuski 915ta key value pairni plot qilolmadim, figsize ga turli xil katta qiymatlarni qo'yib ko'rdim (
plt.figure(figsize=(500,8))
plt.bar(salary_with_specs.keys(),salary_with_specs.values())
plt.savefig('salaries.svg')
```
